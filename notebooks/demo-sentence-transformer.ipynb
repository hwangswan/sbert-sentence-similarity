```python
from sentence_transformers import SentenceTransformer, util
```

## 2. Load pre-trained model

The model we're using is [sentence-transformers/all-MiniLM-L6-v2](https://huggingface.co/sentence-transformers/all-MiniLM-L6-v2)


```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

## 3. Try

Target sentence and list of sentences to perform similarity check

```python
target_sentence = "I'm happy"

sentences = [
    "I'm full of happiness"
]
```

Using SBERT for embedding

```python
embedding = list()
for index in range(len(sentences)):
    embedding.append(model.encode(sentences[index], convert_to_tensor = True))

target_embedding = model.encode(target_sentence, convert_to_tensor = True)
```

Then use Cosine Similarity for evaluation

```python
for index in range(len(embedding)):
    cos_similarity = util.pytorch_cos_sim(target_embedding, embedding[index]).numpy()
    print(f'Similarity score between {target_sentence} and {sentences[index]} is', cos_similarity[0][0])
```
